# Bar plot (numerical-categorical

```python
sns.barplot(titanic['Pclass'],titanic["Age"],hue=titanic['Sex'])
```

```python
sns.boxplot(titanic['Age'],titanic['Sex'],hue=titanic["Survived"])
```

```python
sns.distplot(titanic[titanic["Survived"]==0]['Age'],hist=False)
sns.distplot(titanic[titanic["Survived"]==1]['Age'],hist=False)
```

# HeatMap Categorical-categorical

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean()["Survived"]*100 ##determing the percentage of people died in each class 
```

```python
(titanic.groupby('Pclass').mean()["Survived"]*100).plot(kind='bar')
```

# ClusterMap(cat-cat)

```python
sns.clustermap(pd.crosstab(titanic['Pclass'],titanic['Sex'])) #it tells about the closeness of  gender
```

# Pairplot 

```python
sns.pairplot(iris,hue='species') ##plot each column against each column
```
